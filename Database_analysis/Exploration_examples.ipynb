```python
# EXAMPLE ANALYITICS PAGE


import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("../Data/product_data_db.db")
df = pd.read_sql_query("SELECT * from flavour_table", con)

```

```python
df = df[df['Flavours'] != 'None']
df['Flavours'].value_counts()[0:10]
df['Flavours'].value_counts()[0:10].plot(kind='bar')
```
